```python
x=d[['Age','Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction']]
y=d['Outcome']
```

```python
d.describe()
```

```python
d.isna().sum()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x=scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=40)
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(random_state=42)
dt.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred = dt.predict(x_test)
print(accuracy_score(y_test, y_pred))
```

```python
from sklearn.tree import export_graphviz
import graphviz

feature_cols = ['Age','Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction']

# Generate DOT data string from the decision tree
dot_data = export_graphviz(
    dt,
    feature_names=feature_cols, # Use the correct feature names for the trained model
    filled=True,
    rounded=True,
    special_characters=True
)

# Create a graphviz Source object from the DOT data
graph = graphviz.Source(dot_data)

# Display the graph
graph
```
